```python
import os
from dotenv import load_dotenv
import json
import requests
import re
load_dotenv()
from pandas import json_normalize
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns
import numpy as np
from seaborn import violinplot
```

```python
df_total = pd.read_csv("data/triatlon_merge.csv")
df_total
```

```python
df = df_total
df["Height"] = df["Height"].fillna("Unknown")
df["Weight"] = df["Weight"].fillna("Unknown")
df["Languages"] = df["Languages"].fillna("Unknown")
df["Sponsors"] = df["Sponsors"].fillna("Unknown")
df["Twitter"] = df["Twitter"].fillna("Unknown")
```

```python
lista_nombres = pd.unique(df.name)
lista_nombres
```

```python
pregunta = input(f"Dime que atleta quieres conocer: ")
Index = list(np.where(df["name"] == pregunta))
if pregunta in lista_nombres:
    print(f" Nº ranking: {df.Rank.values[Index]}")
    print(f" Año de nacimiento: {df.YOB.values[Index]}")
    print(f" Pais: {df.Country.values[Index]}")
    print(f" Altura: {df.Height.values[Index]}")
    print(f" Peso: {df.Weight.values[Index]}")
    print(f" Idiomas: {df.Languages.values[Index]}")
    print(f" Sponsor: {df.Sponsors.values[Index]}")
    print(f" Twitter: {df.Twitter.values[Index]}")
    print(f" Carreras ganadas: {df.race_wins.values[Index]}")
    print(f" Podiums: {df.race_podiums.values[Index]}")
else:
    print("El atleta no existe o está mal escrito")
```

### Porcentaje de podiums

```python
sns.set_style("darkgrid")
plot = sns.histplot(x=df_total.race_podium_percentage, kde=True, bins=12)
plot.axvline(x=df_total.race_podium_percentage.mean(), c="red", label="mean")
plot.legend()
```

### Porcentaje de victorias

```python
plot = sns.histplot(x=df_total.race_wins_percentage, kde=True, bins=10)
plot.axvline(x=df_total.race_wins_percentage.mean(), c="red", label="mean")
plot.legend()
```
